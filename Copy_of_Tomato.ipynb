### Feature descriptions:

1. **Average Plant Size (m²)**
   - *Categorical (6 options):* The typical size (area) occupied by a single plant in the field.

   - *Categorical (7 options):* The average number of bees (from type 1) visiting flowers per square meter per minute.

3. **bee_2 Pollination Activity (bees/m²/min)**
   - *Categorical (11 options):* The average number of bees (from type 2) visiting flowers per square meter per minute.

4. **bee_3 Pollination Activity (bees/m²/min)**
   - *Categorical (16 options):* The average number of bees (from type 3) visiting flowers per square meter per minute.

5. **bee_4 Pollination Activity (bees/m²/min)**
   - *Categorical (14 options):* The average number of bees (from type 4) visiting flowers per square meter per minute.

6. **Maximum Upper Bloom Temperature (°C)**
   - *Categorical (8 options):* The highest recorded daily air temperature within the optimal range during the bloom season.

7. **Minimum Upper Bloom Temperature (°C)**
   - *Categorical (5 options):* The lowest recorded daily air temperature within the optimal range during the bloom season.

8. **Average Upper Bloom Temperature (°C)**
   - *Categorical (6 options):* The average daily air temperature within the optimal range during the bloom season.

9. **Maximum Lower Bloom Temperature (°C)**
   - *Categorical (6 options):* The highest recorded daily air temperature below the optimal range during the bloom season.

10. **Minimum Lower Bloom Temperature (°C)**
    - *Categorical (7 options):* The lowest recorded daily air temperature below the optimal range during the bloom season.

11. **Average Lower Bloom Temperature (°C)**
    - *Categorical (5 options):* The average daily air temperature below the optimal range during the bloom season.

12. **Total Rainy Days**
    - *Categorical (6 options):* The total number of days with measurable precipitation (greater than zero) during the bloom season.

13. **Proportion of Rainy Days**
    - *Categorical (8 options):* The average number of days with measurable precipitation per day during the bloom season (a value between 0 and 1).

14. **Fruit Set**
    - *Numerical (1525 options):* The number of plants that have set fruit.

15. **Mass of Fruit**
    - *Numerical (1515 options):* The mass (weight) of the harvested fruit.

16. **Average Seeds**
    - *Numerical (2066 options):* The average number of seeds per fruit.

17. **Tomato Yield**
    - *Numerical (776 options):* The total yield of tomatoes harvested.

```python
categorical_columns = [
    'Average Plant Size',
    'Bee_1 Pollination Activity',
    'bee_2 Pollination Activity',
    'bee_3 Pollination Activity',
    'bee_4 Pollination Activity',
    'Maximum Upper Bloom Temperature',
    'Minimum Upper Bloom Temperature',
    'Average Upper Bloom Temperature',
    'Maximum Lower Bloom Temperature',
    'Minimum Lower Bloom Temperature',
    'Average Lower Bloom Temperature',
    'Total Rainy Days',
    'Proportion of Rainy Days'
]

for column in categorical_columns:
    print(f"{column}: {', '.join(map(str, tomatoes[column].unique()))}\n")
```

**conclusion**:

lets correct it and change the missing value to NaN
then we can convert all this column to float

```python
tomatoes.loc[tomatoes['Average Upper Bloom Temperature'] == 'missing', 'Average Upper Bloom Temperature'] = np.NaN
tomatoes['Average Upper Bloom Temperature'] = tomatoes['Average Upper Bloom Temperature'].astype(float)
```

```python
numerical_columns = [
    'fruit_set',
    'mass_of_fruit',
    'average_seeds',
    'tomato_yield',
]

for column in numerical_columns:
    string_values = tomatoes[column][tomatoes[column].apply(lambda x: isinstance(x, str))]
    if not string_values.empty:
        print(f"{column}: string values - {', '.join(map(str, string_values.unique()))}")

    # Check for negative values or zeros
    negatives_zeros = tomatoes[column][~tomatoes[column].apply(lambda x: isinstance(x, str)) & (tomatoes[column] <= 0)]
    if not negatives_zeros.empty:
        print(f"{column}: Zero / Negative values - {', '.join(map(str, string_values.unique()))}")
```

### Tomato yeild

so lets split it to categorial by range:
1000-2000, 2001-3000, 3001-4000, 4001-5000, 5001-6000, 6001-7000, 7001-8000, 8001-9000

```python
# Define bin edges
bin_edges = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000]

# Define labels
labels = ['1000-2000', '2001-3000', '3001-4000', '4001-5000', '5001-6000', '6001-7000', '7001-8000', '8001-9000']

# Create new categorical column
tomatoes['yield_range'] = pd.cut(tomatoes['tomato_yield'], bins=bin_edges, labels=labels)

# Count values in each group and specify the order of categories
group_counts = tomatoes['yield_range'].value_counts().loc[labels]

# Plot bar plot
group_counts.plot(kind='bar')
plt.xlabel('Yield Range')
plt.ylabel('Count')
plt.title('Distribution of Tomato Yield Ranges')
plt.show()
```

lets also split the data to 3 categorial low-yield, mid-yield and max-yeild by range of 0-3000, 3001-6000, 6001-9000

```python
# Define bin edges
bin_edges = [0, 3000, 6000, 9000]

# Define labels
labels = ['low-yield', 'mid-yield', 'max-yield']

# Create new categorical column
tomatoes['yield_category'] = pd.cut(tomatoes['tomato_yield'], bins=bin_edges, labels=labels)

# Count values in each group and specify the order of categories
group_counts = tomatoes['yield_category'].value_counts()

# Plot bar plot
group_counts.plot(kind='bar')
plt.xlabel('Yield Category')
plt.ylabel('Count')
plt.title('Distribution of Tomato Yield Categories')
plt.show()

```

#### Conclusion:

We divided the yield into two categorical fields: one by a range of 1000 and the other by a range of 3000, representing minimum, medium, and maximum yield categories.

### Average Plant Size

lets check the distribution of average plant size

```python
# Calculate the percentage distribution
percentage_distribution = (tomatoes['Average Plant Size'].value_counts(normalize=True) * 100).round(2)
count_distribution = tomatoes['Average Plant Size'].value_counts()

# Plot a bar chart for 'Average Plant Size'
ax = percentage_distribution.plot(kind='bar')

# Add labels and title
plt.xlabel('Average Plant Size')
plt.ylabel('Percentage')
plt.title('Percentage Distribution of Average Plant Size')

# Add '%' to the y-axis labels and count values below each bar
for i, p in enumerate(ax.patches):
    ax.annotate(f'{p.get_height()}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 15), textcoords='offset points')
    ax.annotate(f'n={count_distribution[percentage_distribution.index[i]]}',
                (p.get_x() + p.get_width() / 2.,  p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points', color='red')

# Adjust y-axis limits
plt.ylim(top=ax.get_ylim()[1] * 1.1)

# Show the plot
plt.show()
```

 we can see that almost all of the values are 25.0 or 12.5.

 2% are from other values

 so lets create new column call 'average_plant_size_bellow_20' and make it boolean

```python
tomatoes['average_plant_size_bellow_20'] = tomatoes['Average Plant Size'] <= 20.0
```

lets check how it affect the tomato_yeild

```python
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Plot bar plot for yield range
grouped = tomatoes.groupby(['yield_range', 'average_plant_size_bellow_20']).size().unstack()
grouped['Total'] = grouped[True] + grouped[False]
grouped['True_Percentage'] = (grouped[True] / grouped['Total']) * 100
grouped['False_Percentage'] = (grouped[False] / grouped['Total']) * 100
grouped[['True_Percentage', 'False_Percentage']].plot(kind='bar', ax=axes[0], color=['lightgreen', 'lightcoral'])
axes[0].set_xlabel('Yield Range')
axes[0].set_ylabel('Percentage')
axes[0].set_title('Percentage Distribution of Average Plant Size Below 20 by Yield Range')
axes[0].yaxis.set_major_formatter(plt.FuncFormatter('{:.0f}%'.format))
axes[0].legend(title='Average Plant Size Below 20', loc='upper left', bbox_to_anchor=(1, 1))

# Plot bar plot for yield category
grouped = tomatoes.groupby(['yield_category', 'average_plant_size_bellow_20']).size().unstack()
grouped['Total'] = grouped[True] + grouped[False]
grouped['True_Percentage'] = (grouped[True] / grouped['Total']) * 100
grouped['False_Percentage'] = (grouped[False] / grouped['Total']) * 100
grouped[['True_Percentage', 'False_Percentage']].plot(kind='bar', ax=axes[1], color=['lightgreen', 'lightcoral'])
axes[1].set_xlabel('Yield Category')
axes[1].set_ylabel('Percentage')
axes[1].set_title('Percentage Distribution of Average Plant Size Below 20 by Yield Category')
axes[1].yaxis.set_major_formatter(plt.FuncFormatter('{:.0f}%'.format))
axes[1].legend(title='Average Plant Size Below 20', loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

We can infer from the graph that when the average plant size is below 20 m², the yield increases, and vice versa

```python
# Define the columns
pollination_columns = ['bee_1 Pollination Activity',
                       'bee_2 Pollination Activity',
                       'bee_3 Pollination Activity',
                       'bee_4 Pollination Activity']

# Create a 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(20, 10))

# Iterate through the columns and plot the bar plot in each subplot
for i, column in enumerate(pollination_columns):
    row = i // 2
    col = i % 2
    # Calculate the percentage distribution
    percentage_distribution = (tomatoes[column].value_counts(normalize=True) * 100).round(2)
    count_distribution = tomatoes[column].value_counts()

    # Sort the index (x-axis labels)
    percentage_distribution = percentage_distribution.sort_index()

    # Plot a bar chart for column
    ax = percentage_distribution.plot(kind='bar', ax=axes[row, col], color='skyblue')

    # Add labels and title
    ax.set_xlabel(column)
    ax.set_ylabel('Percentage')
    ax.set_title('Percentage Distribution of ' + column)

    # Add '%' to the y-axis labels and count values below each bar
    for j, p in enumerate(ax.patches):
        ax.annotate(f'{p.get_height()}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 15), textcoords='offset points')
        ax.annotate(f'n={count_distribution[percentage_distribution.index[j]]}',
                    (p.get_x() + p.get_width() / 2.,  p.get_height()),
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points', color='red')

    # Adjust y-axis limits
    ax.set_ylim(top=ax.get_ylim()[1] * 1.1)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

It can be seen that for each type of pollination activity, there are only a few single observations that have a frequency higher than 1 percent.

#### Average Tomato yield for each pollination activity

```python
# Create a single plot
plt.figure(figsize=(10, 6))

# Iterate through each bee_X
for bee in range(1, 5):
    bee_col = f'bee_{bee} Pollination Activity'

    # Filter out pollination values greater than 1 and with less than 20 observations
    filtered_data = tomatoes[tomatoes[bee_col].between(0, 1) & (tomatoes.groupby(bee_col)['tomato_yield'].transform('count') >= 5)]

    # Group by pollination activity and calculate average tomato yield
    avg_yield_by_pollination = filtered_data.groupby(bee_col)['tomato_yield'].mean()

    # Create line plot
    plt.plot(avg_yield_by_pollination.index, avg_yield_by_pollination.values, marker='o', label=bee_col, alpha=0.7)

# Add labels and title
plt.xlabel('Pollination Activity')
plt.ylabel('Average Tomato Yield')
plt.title('Average Tomato Yield vs Pollination Activity')
plt.legend()

# Set y-axis limits
plt.ylim(4000, 7000)

# Adjust layout
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, 1)
plt.show()
```

The graph illustrates the relationship between different levels of pollination activity by all four types of bees and the average yield of tomatoes.
Each line represents the average tomato yield corresponding to the level of pollination activity of a specific bee species.
In this graph, we reduced the pollination activity above 1, since the number is extreme and is of low frequency, and we also reduced occasional points with less than 5 observations.

It is possible to confirm several conclusions we received based on the previous graphs for the groups.
For type 1 bees:
The higher the amount of pollination activity of the bees, the lower the yield of the tomatoes.
For type 2-4 bees:
The higher the amount of pollination activity of bees, the higher the average yield of tomatoes.

##### Bee type 2 to 4:

```python
# Create a new column with the sum of bee pollination activity for bee_2 to bee_4
tomatoes['bee_2_to_4_pollination_activity_total'] = tomatoes[
    ['bee_2 Pollination Activity',
     'bee_3 Pollination Activity',
     'bee_4 Pollination Activity']].sum(axis=1)

# Filter out observations with less than 5 for bee_2 to bee_4 sum
filtered_data = tomatoes.groupby('bee_2_to_4_pollination_activity_total').filter(lambda x: len(x) >= 20)

# Calculate average bee pollination activity
avg_bee_pollination = filtered_data.groupby('bee_2_to_4_pollination_activity_total')['tomato_yield'].mean()

# Plot a line graph
plt.figure(figsize=(10, 6))
plt.plot(avg_bee_pollination.index, avg_bee_pollination.values, marker='o', linestyle='-', color='gray')
plt.xlabel('Sum of Bee Pollination Activity (bee_2 to bee_4)')
plt.ylabel('Average Tomato yield')
plt.title('Average tomato yield vs Sum of Bee Pollination Activity (bee_2 to bee_4)')
plt.grid(True)

# Add x values on each point
for x, y in zip(avg_bee_pollination.index, avg_bee_pollination.values):
    plt.text(x, y + 50, f'{x:.2f}', ha='center', va='bottom', fontsize=8, color='red')


plt.show()
```

```python
# List of optimal average temperatures
rainy_days_values = list(rainy_days_groups.values())

# Create an empty DataFrame
df = pd.DataFrame(index=range(0, 8000, 250), columns=[rainy_days_value for rainy_days_value in rainy_days_values])

# Fill the DataFrame with counts
for rainy_days_value in rainy_days_values:
    rainy_days_data = tomatoes[tomatoes['total_rainy_days_groups'] == rainy_days_value]
    rainy_days_counts = rainy_days_data['tomato_yield'].value_counts(bins=range(0, 8001, 250)).sort_index()
    df[rainy_days_value] = rainy_days_counts

# Plot line graph with light colors
df.plot(figsize=(12, 8), color=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'])

# Add labels and title
plt.xlabel('Tomato Yield')
plt.ylabel('Frequency')
plt.title('Distribution of Tomato Yield by Number of Rainy Days')
plt.grid(True)
plt.legend(title='Number of Rainy Days')

# Set x-axis ticks
plt.xticks(range(0, 8001, 1000))

# Show the plot
plt.show()
```

You can see that a few days 1 and 16 give more high output of 6000+, in 16 days.

We will try to look at the categories we created for the tomato yield and see the impact more clearly.

```python
def plot_bar_by_group(data, group_column, group_values, xlabel, ylabel, title, rainy_days_values):
    # Filter data for group values
    data_filtered = data[data[group_column].isin(group_values)]

    # Create a DataFrame with count of observations for each group and average temperature
    df = data_filtered.groupby([group_column, 'total_rainy_days_groups']).size().unstack()

    # Plot the bar plot
    plt.figure(figsize=(12, 4))
    width = 0.15
    colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightpink']
    for i, rainy_days in enumerate(rainy_days_values):
        x = range(len(group_values))
        y = df[rainy_days].values
        plt.bar([pos + width * i for pos in x], y, width=width, label=rainy_days, color=colors[i])

    # Add labels and title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks([pos + 1.5 * width for pos in x], group_values)
    plt.grid(True)
    plt.legend(title='Average Upper Bloom Temperature', loc='upper left')

    # Show the plot
    plt.tight_layout()
    plt.show()


# Call the function for yield_range
plot_bar_by_group(tomatoes, 'yield_range', ['1000-2000', '2001-3000', '3001-4000', '4001-5000', '5001-6000', '6001-7000', '7001-8000', '8001-9000'],
                  xlabel='Yield Range', ylabel='Frequency', title='Frequency of Average Upper Bloom Temperature by Yield Range', rainy_days_values=rainy_days_values)

# Call the function for yield_category
plot_bar_by_group(tomatoes, 'yield_category', ['low-yield', 'mid-yield', 'max-yield'],
                  xlabel='Yield Category', ylabel='Frequency', title='Frequency of Average Upper Bloom Temperature by Yield Category', rainy_days_values=rainy_days_values)
```

```python
# Group by 'total_rainy_days_groups' and calculate the mean of 'tomato_yield'
result = tomatoes.groupby('total_rainy_days_groups')['tomato_yield'].agg(['mean', 'count'])

# Plotting
ax = result['mean'].plot(kind='bar', color='lightblue')
plt.xlabel('Total Rainy Days Groups')
plt.ylabel('Average Tomato Yield')
plt.title('Average Tomato Yield by Total Rainy Days Groups')
plt.xticks(rotation=45)

# Adding total number on top of each bar
for i, v in enumerate(result['mean']):
    ax.text(i, v + 100, str(round(v,1)), color='red', ha='center')

# Adjusting y-label
max_value = result['mean'].max()
plt.yticks(range(0, int(max_value) + 2000, 1000))

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

From these graphs you can clearly see that a quantity of 34 days gives the most average output followed by 24 days, then 16 and then one day.
In addition, it can be seen with the help of the first and second graphs that the amount of just one rainy day gives the highest maximum yield (the most observations over 7000 tomato yield units).
followed by 16 days of rain.

The main conclusion we can draw from this is that the lower the number of rainy days (the lowest measured -1), the higher the yield seen in the results.
